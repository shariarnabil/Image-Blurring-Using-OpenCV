```python
#importing...
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Allow the user to upload an image file
# Using Google Colab's 'files.upload()' function to upload the image
from google.colab import files
uploaded = files.upload()  # Upload the image file here
```

```python
# Load the uploaded image
# The uploaded file is accessed by its name (list of uploaded file names)
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path)
```

```python
# Convert BGR to RGB for displaying with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Display the original image
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
 #Blur Image using cv2.blur() with Kernel 3×3
img_blur_3 = cv2.blur(img, (3, 3))

# Blur Image using cv2.blur() with Kernel 50×50
img_blur_50 = cv2.blur(img, (50, 50))

# Blur Image using cv2.boxFilter() with Kernel 3×3
img_b_blur_3 = cv2.boxFilter(img, -1, (3, 3))

# Blur Image using cv2.boxFilter() with Kernel 50×50
img_b_blur_50 = cv2.boxFilter(img, -1, (50, 50))
```

```python
# Resize images for easier visualization
img_blur_3 = cv2.resize(img_blur_3, (650, 400))
img_blur_50 = cv2.resize(img_blur_50, (650, 400))
img_b_blur_3 = cv2.resize(img_b_blur_3, (650, 400))
img_b_blur_50 = cv2.resize(img_b_blur_50, (650, 400))
```

```python
# Add text to each image
```

```python
cv2.putText(img_blur_3, "blur 3x3", (10, 50), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 0, 55), 3)
```

```python
cv2.putText(img_blur_50, "blur 50x50", (10, 50), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 0, 55), 3)
```

```python
cv2.putText(img_b_blur_3, "box blur 3x3", (10, 50), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 0), 3)
```

```python
cv2.putText(img_b_blur_50, "box blur 50x50", (10, 50), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 0), 3)
```

```python
# Stack the blurred images horizontally and vertically for comparison
# Horizontal stacking for normal and box blur, and vertical stacking to combine them all
blurr_2_img = np.hstack((img_blur_3, img_blur_50))  # Horizontal stacking of average blur images
filter_blur_2_img = np.hstack((img_b_blur_3, img_b_blur_50))  # Horizontal stacking of box blur images
img_4 = np.vstack((blurr_2_img, filter_blur_2_img))  # Vertical stacking of both horizontal stacks
```

```python
# Convert BGR to RGB for displaying with matplotlib
img_4_rgb = cv2.cvtColor(img_4, cv2.COLOR_BGR2RGB)
```

```python
# Display the final image that shows all the different blurring techniques
plt.figure(figsize=(10, 10))  # Define figure size for better visualization
plt.imshow(img_4_rgb)
plt.title("Show 4 Blur Images")  # Add a title to describe the final image
plt.axis('off')  # Turn off axis for better presentation
plt.show()  # Display the combined image showing all blur effects
```
